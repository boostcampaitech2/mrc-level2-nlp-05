```python
import os
import json

import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
OUTPUT_DIR = "/opt/ml/code/outputs/klue_roberta_large2"
N = 5

with open(os.path.join(OUTPUT_DIR, "predictions.json"), "r") as f:
    pred = json.load(f)
pred = [[k, v] for k, v in pred.items()]

with open(os.path.join(OUTPUT_DIR, "nbest_predictions.json"), "r") as f:
    nbest = json.load(f)
nbest = [[k, v] for k, v in nbest.items()]
```

```python
full_list = []
for item in nbest:
    item_out = [item[0]]
    for i in range(N):
        try:
            ith_pred = item[1][i]
            for k, v in ith_pred.items():
                item_out.append(v)
        except:
            item_out.extend([-1e9, -1e9, "", 0])
    full_list.append(item_out)
```

```python
pred = pd.DataFrame(pred)
nbest = pd.DataFrame(full_list)

new_column_name = ["data_id"]
for i in range(N):
    new_column_name.append("start_logit_" + str(i+1))
    new_column_name.append("end_logit_" + str(i+1))
    new_column_name.append("text_" + str(i+1))
    new_column_name.append("probability_" + str(i+1))
nbest.columns = new_column_name
```
